```python
domc = pd.DataFrame({
    'Image name': [],
    'Dominant colour percentage': [],
    'Color': []
})
```

```python
data = []

for i in range(1,11):
    
    print '****************************** IMAGE %s *************************************** ' %i
    
    image_name = 'image ' + str(i) + '.jpg'
    
    image = cv2.imread(image_name)
    
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    print 'Image shape: '+str(image.shape)
    
    '''
    plt.figure(i)
    plt.axis("off")
    plt.imshow(image)
    '''
    image = image.reshape((image.shape[0]*image.shape[1],3))
    print 'Shape of image after image reshape: '+str(image.shape)

    kmeans = KMeans(n_clusters=3, random_state=0).fit(image)

    label = [0,0,0]

    for x in kmeans.labels_:
        label[x] += 1

    label_sum = image.shape[0]

    label = [round((float(x)/label_sum)*100, 2) for x in label]
    m = max(label)
    y = label.index(m)
    print 'Percentage of different colours: '+str(label)

    centroids = [x.astype('uint8').tolist() for x in kmeans.cluster_centers_]
    print centroids
    
    clr = [webcolors.rgb_to_hex(c) for c in centroids]
    print 'colours hex codes: '+ str(clr)
    dc = clr[y]

    plt.pie(label, colors=clr, shadow=True, autopct='%1.1f%%')
    plt.axis('equal')
    plt.show()
    
    l = []
    l.append('Image'+str(i))
    l.append(m)
    l.append(dc)
    
    data.append(l)
    
    domc = domc.append(pd.DataFrame({
        'Image name' : [data[i-1][0]],
        'Dominant colour percentage' : [data[i-1][1]],
        'Color' : [data[i-1][2]]
    }))
```

```python
print domc
```

```python
sns.barplot(data=domc, x='Image name', y='Dominant colour percentage')
```
